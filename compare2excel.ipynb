```python
import pandas as pd

# Assuming you have pandas DataFrame df1 and df2

result_data = []  # Create an empty list to store the results

for i in range(0, 44700):
    s1_value = df1['ptx_parcel_number'][i]
    found_rows_df2 = df2[df2['ptx_parcel_number'] == s1_value]

    if not found_rows_df2.empty:  # Check if there are matching rows in df2
        s2 = (df1['assr_parcel_use_code'][i] == found_rows_df2['assr_parcel_use_code'].values[0])

        if not s2:
            s5 = found_rows_df2['assr_parcel_use_code'].values[0] == 'nan'
            if s5:
                result_data.append([df1['ptx_parcel_number'][i], df1['assr_parcel_use_code'][i], found_rows_df2['assr_parcel_use_code'].values[0]])

# Create a DataFrame from the result_data list
result_df = pd.DataFrame(result_data, columns=['ptx_parcel_number', 'assr_parcel_use_code_df1', 'assr_parcel_use_code_df2'])

# Save the DataFrame to an Excel file
result_df.to_excel('output_results.xlsx', index=False)
```
